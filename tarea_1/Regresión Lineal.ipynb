## 1. Generación del conjunto de datos

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generar datos con distribución normal
np.random.seed(0)  # Para reproducibilidad
num_datos = 300
X = 2 * np.random.rand(num_datos, 1)
Y = 4 + 3 * X + np.random.randn(num_datos, 1)
print("La longitud del conjunto de datos es: ", len(X))
```

## 2. Graficar datos

```python
# Graficar los datos
plt.scatter(X, Y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Dataset de distribución normal')
plt.show()
```

## 3. Modificación de datos

```python
# Convertir X a una matriz de tamaño (num_datos, 1) para poder usarlo en Scikit-Learn
X = X.reshape(-1, 1)
```

## 4. Construcción del modelo

```python
# Crear un modelo de regresión lineal
modelo = LinearRegression()
```

```python
# Entrenar el modelo
modelo.fit(X, Y)
```

```python
# Imprimir coeficientes del modelo
print('Coeficiente de la pendiente:', modelo.coef_[0])
print('Término independiente:', modelo.intercept_)
```

```python
# Plotear los datos y la regresión lineal
plt.scatter(X, Y, color='blue', label='Datos')
plt.plot(X, modelo.predict(X), color='red', linewidth=2, label='Regresión Lineal')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Regresión Lineal con Dataset de distribución normal')
plt.legend()
plt.show()
```
